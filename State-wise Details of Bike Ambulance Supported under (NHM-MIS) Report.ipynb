```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("State-wise Details of Bike Ambulance Supported under National Health Mission (NHM-MIS) Report during September 2022.csv")
```

```python
df=df.fillna(0)
```

```python
df_cleaned = df[df["States/UTs"] != "Total"].copy()
```

```python
df_cleaned["Bike Ambulance"] = pd.to_numeric(df_cleaned["Bike Ambulance"], errors="coerce")
```

```python
df_cleaned = df_cleaned.astype({"Bike Ambulance": "int64"})
```

```python
df_cleaned = df_cleaned.sort_values("Bike Ambulance", ascending=True)
```

```python
df_cleaned = df_cleaned.fillna(0)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df_cleaned["States/UTs"], df_cleaned["Bike Ambulance"], color='skyblue')
plt.xlabel("States/UTs")
plt.ylabel("Number of Bike Ambulances")
plt.title("Bike Ambulances Distribution Across States")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.barh(df_cleaned["States/UTs"], df_cleaned["Bike Ambulance"], color='green')
plt.xlabel("Number of Bike Ambulances")
plt.ylabel("States/UTs")
plt.title("Bike Ambulances Distribution Across States")
plt.gca().invert_yaxis()  
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df_cleaned["States/UTs"], df_cleaned["Bike Ambulance"], marker='o', linestyle='-', color='purple')
plt.xlabel("States/UTs")
plt.ylabel("Number of Bike Ambulances")
plt.title("Trend of Bike Ambulances Across States")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned["States/UTs"], df_cleaned["Bike Ambulance"], color='red', marker='o')
plt.xlabel("States/UTs")
plt.ylabel("Number of Bike Ambulances")
plt.title("Scatter Plot of Bike Ambulances Across States")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned["Bike Ambulance"], bins=5, color='cyan', edgecolor='black')
plt.xlabel("Number of Bike Ambulances")
plt.ylabel("Frequency")
plt.title("Distribution of Bike Ambulances")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.boxplot(df_cleaned["Bike Ambulance"], vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.ylabel("Number of Bike Ambulances")
plt.title("Box Plot of Bike Ambulances Across States")
plt.show()
```

```python
sns.pairplot(df_cleaned)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df_cleaned.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df_cleaned["Other Ambulances"] = [20, 50, 40, 60]  
df_cleaned.set_index("States/UTs")[["Bike Ambulance", "Other Ambulances"]].plot(kind="bar", stacked=True, figsize=(8, 6), color=["blue", "orange"])
plt.xlabel("States/UTs")
plt.ylabel("Number of Ambulances")
plt.title("Stacked Bar Chart: Bike vs. Other Ambulances")
plt.xticks(rotation=45)
plt.legend(title="Ambulance Type")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.fill_between(df_cleaned["States/UTs"], df_cleaned["Bike Ambulance"], color="lightcoral", alpha=0.6)
plt.xlabel("States/UTs")
plt.ylabel("Number of Bike Ambulances")
plt.title("Cumulative Bike Ambulances Across States")
plt.xticks(rotation=45)
plt.show()
```
